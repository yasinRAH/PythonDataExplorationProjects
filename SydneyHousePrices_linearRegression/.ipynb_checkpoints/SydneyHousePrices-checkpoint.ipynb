```python
#find averages per area
df_avgPerArea = df.groupby(by="suburb").mean()
df_avgPerArea
```

```python
#put average sell price per area into new column of df
df["MeanSellPricePerArea"] = df.groupby(by="suburb")["sellPrice"].transform("mean")
df
```
